## Plotting Lesson

```python
import os
import numpy as stumpy
import pylab as py
```

```python
distance_file=os.path.join('distance_data_headers.csv')
distances=stumpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
headers=distances[0,:]
data=distances[1:,:]
data=data.astype(stumpy.float)
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
sample=headers[1]
sample2=headers[2]
plt.plot(data[:,1], label=sample, color='b')
plt.plot(data[:,2], label=sample2, color='c')
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstrom)')
plt.title('Cool Graph')
plt.legend()

plt.savefig(F'{sample}.png', dpi=300)
```

```python
data_size=len(data[0,:])

for num in range(1,data_size):
    sample=headers[num]
    plt.plot(data[:,num], label=sample)
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstrom)')
plt.title('Cool Graph')
plt.legend()
```

```python
data_size=len(data[0,:])
color=[0, 'c', 'm', 'g', 'r']

for num in range(1,data_size):
    plt.figure()
    sample=headers[num]
    plt.plot(data[:,num], label=sample, color=color[num])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (Angstrom)')
    plt.title('Cool Graph')
    plt.legend()
    plt.savefig(F'{sample}.png', dpi=300)
```
